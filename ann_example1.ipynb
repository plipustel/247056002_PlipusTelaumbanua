### 1A. PEMISAHAHAN DUA KELOMPOK (WX + b) = 0

```python
""""
TUGAS 2 :   Artificial Neural Network
NAMA    :   Plipus Telaumbanua
NIM     :   247056002
"""
import numpy as np
import matplotlib.pyplot as plt

# Dataset
X = np.array([
    [1, 1],
    [1, 0],
    [0, 1],
    [0, 0]
])
y = np.array([1, -1, -1, -1])

# Memisahkan kelompok berdasarkan target
X_target_1 = X[y == 1]
X_target_neg1 = X[y == -1]

# Visualisasi awal data
plt.scatter(X_target_1[:, 0], X_target_1[:, 1], color='blue', marker='o', label='Target 1')
plt.scatter(X_target_neg1[:, 0], X_target_neg1[:, 1], color='red', marker='s', label='Target -1')
plt.xlabel('X1')
plt.ylabel('X2')
plt.legend()
plt.title('Visualisasi Data Berdasarkan Target')
plt.show()
```

### 2B. NILAI W1, W2, W3 AGAR LINE BOUNDRY OPTIMAL

```python
# Inisialisasi bobot dan bias
w = np.zeros(3)  # [w1, w2, w3] dengan w3 sebagai bias
learning_rate = 0.1

# Menambahkan bias 1 di setiap data
X_with_bias = np.hstack((X, np.ones((X.shape[0], 1))))

# Fungsi aktivasi untuk perceptron
def activation(x):
    return 1 if x >= 0 else -1

# Perceptron training
for epoch in range(10):
    for i in range(len(X_with_bias)):
        prediction = activation(np.dot(X_with_bias[i], w))
        error = y[i] - prediction
        w += learning_rate * error * X_with_bias[i]

print("Nilai w1, w2, dan w3 yang optimal:", w)
```

### 3B. BOUNDRY DECISION f(x1, x2)

```python
# Fungsi f(x1, x2) untuk decision boundary
def f(x1, x2):
    return -(w[0] * x1 + w[2]) / w[1]  # Menggunakan w[2] sebagai bias

# Visualisasi decision boundary
x_values = np.linspace(-0.5, 1.5, 100)
y_values = f(x_values, 0)  # Menentukan y berdasarkan fungsi f(x1, x2)

plt.scatter(X_target_1[:, 0], X_target_1[:, 1], color='blue', marker='o', label='Target 1')
plt.scatter(X_target_neg1[:, 0], X_target_neg1[:, 1], color='red', marker='s', label='Target -1')
plt.plot(x_values, y_values, color='green', label='Decision Boundary')
plt.xlabel('X1')
plt.ylabel('X2')
plt.legend()
plt.title('Visualisasi Decision Boundary')
plt.show()
```

### 3C.THHRESHOLD = 0.8

```python
import pandas as pd

# Data Input sesuai tabel pada Step 4
data = {
    'Inputs': ['(1, 1)', '(1, 0)', '(0, 1)', '(0, 0)'],
    'Weight': ['50%', '40%', '40%', '50%'],
    'Weighted Sum': [1.5, 0.8, 0.8, 0.5]
}

# Menghitung Output Produced berdasarkan threshold 0.8
threshold = 0.8
data['Output Produced'] = ['1' if ws >= threshold else '-1' for ws in data['Weighted Sum']]

# Target Output berdasarkan tabel yang diberikan
data['Actual Output'] = ['1', '-1', '-1', '-1']

# Menentukan Action berdasarkan perbandingan antara Output Produced dan Actual Output
data['Action'] = [
    'None' if data['Output Produced'][i] == data['Actual Output'][i] else 'Decresase'
    for i in range(len(data['Output Produced']))
]

# Membuat DataFrame untuk visualisasi dalam bentuk tabel
df = pd.DataFrame(data)

# Menampilkan tabel
print(df)
```
